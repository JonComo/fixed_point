```python
import numpy as np
import window

w = window.Window(width=800, height=800)
w.set_fps(60)

POINTS = 2000

points = np.random.randn(POINTS, 2)

W1 = np.random.randn(2, 20)
W2 = np.random.randn(20, 2)

#t = 0.1
#W = np.array([[np.cos(t), -np.sin(t)],
#              [np.sin(t), np.cos(t)]])

#W[0, 1] *= -0.3

w.set_particles(points)

act = lambda x: 1 / (1.0 + np.exp(-x))
actp = lambda x: act(x) * (1-act(x))

idx = 0

zoom = 200.0

noisy = False
clear = True

def screen_to(p):
    return (np.array(p) - (w.width/2, w.height/2))/zoom

def to_screen(p):
    return p * zoom + (w.width/2, w.height/2)

while True:
    w.clock.tick()
    if clear:
        w.clear()
    
    if w.mouse_pressed:
        for i in range(10):
            p = screen_to(w.mouse) + np.random.randn(1, 2) * 0.05
            points[idx] = p
            idx += 1
            if idx >= points.shape[0]:
                idx = 0
                
    if w.pressed(window.key.SPACE):
        noisy = not noisy
        
    if w.pressed(window.key.C):
        clear = not clear
                
    if w.pressed(window.key.R):
        w.clear()
        points = np.random.randn(POINTS, 2)
        W1 = np.random.randn(2, 4)
        W2 = np.random.randn(4, 2)
        
    if noisy:
        i = np.random.randint(0, points.shape[0]-10)
        for j in range(10):
            points[i+j] = np.random.randn(1, 2)
        
    w.reset_keys()
    
    w.update_particles(to_screen(points))
    
    #points = points.dot(W)
    
    points = np.tanh(points.dot(W1))
    points = np.tanh(points.dot(W2))
    
    if w.update() == False:
        break
    
w.close()
```
